### Question 2
#### We want to numerically analyse the Atri model for intercellular calcium dynamics given by the following equations:

$$\mathbf{\frac{d c}{d t}=\mu k_f h \left(b+ \frac{V_1c}{k_1+c} \right)-\frac{\gamma c}{k_{\gamma}+c}+\beta}$$

$$\mathbf{\tau_h \frac{d h}{d t}=\frac{k_2^2}{k_2^2+c^2}-h}$$

#### We treat $\mathbf{\mu}$ as the main bifurcation parameter and set the other parameters as suggested in the question. First we want to plot the phase portraits for different values of the parameter $\mathbf{\mu}$ to show different dynamics. Due to the scientific context (c and $\mathbf{\mu}$ are concentrations) we only consider positive input parameters for c and $\mathbf{\mu}$. We use the functions "phaseplot" and "colourlines" which I wrote to plot phase portraits. The code and a detailed description of these functions can be found in the appendix. The plots created with "phaseplot" are meant to give an overview of the dynamics and the direction. The function has a "cut" option, meaning if "cut" is activated we only consider the last 10% of the steps which allows us to plot the object the system is attracted to. The function "colourlines" helps us to plot a denser phase portrait with a lot of trajectories. 

```matlab
phaseplot(0.2,0)
box on;
xlabel('c','FontSize',20);
ylabel('h','FontSize',20);
set(gca,'FontSize',16);
t=title('Phase portrait for \mu = 0.2');
set(t,'FontSize',20);
set(gca,'Xlim',[0,3])
set(gca,'Ylim',[0,2])
```

#### For $\mathbf{\mu}$=0.2  the system is attracted to a stable fixed point which is around (0,1), i.e. the $\mathbf{Ca^{2+}}$ concentration is zero. Playing around with the parameters, we see that the system has this stable fixed point for $\mathbf{\mu}$ between 0 and circa 0.5 (considering $\mathbf{\mu}$ in steps of 0.1). To get a better overview of the convergence we plot a dense phase portrait.

```matlab
colourlines(0.2,0)
xlabel('c','FontSize',20);
ylabel('h','FontSize',20);
set(gca,'FontSize',16);
t=title('Dense phase portrait for \mu = 0.2');
set(t,'FontSize',20);
set(gca,'Xlim',[0,4])
set(gca,'Ylim',[0,2])
```

#### To better understand our experimental results about the systems' dynamics we plot a bifurcation diagram where $\mathbf{\mu}$ is the bifurcation parameter. Confirming our results from before, we see that the system has a stable fixed point for $\mathbf{\mu}$ between 0 and circa 0.58. The fixed point then becomes unstable and the system is attracted to a limit cycle for $\mathbf{\mu}$ between circa 0.58 and circa 0.98. For $\mathbf{\mu}$ between circa 0.98 and 1.5 the system is attracted to a stable fixed point again. What we didn't find out before is that the limit cycle is unstable for particular values of $\mathbf{\mu}$. We can see the stability in the coloured plot. We have two hopf bifurcations (marked (1) and (2) in the plot.). To classify the hopf bifucations we look at the coloured plot which additionally indicates the stabilities. Suppose we start the system in the state $\mathbf{x^*}$=a, where a is the stable fixed point the system is attracted to for $\mathbf{\mu}$ small, and then slowly increase the parameter $\mathbf{\mu}$. The state remeins in a until we reach the first hopf bifurcation point where a becomes unstable. The state then suddenly jumps up to the stable state of the limit cycle and stays on there until the limit cycle becomes unstable (second hopf bifurcation point). Then it jumps back. These jumps are characteristic for subcritical hopf bifurcations. Therefore we can classify both hopf bifurcations as subcritical.

### Question 3

#### We consider a one dimensional map defined by
 
 $$\mathbf{x_{n+1}=\exp\left(-ax^2_n \right) +b}$$
 
#### where a and b are real numbers. We observe that the function is a Gaussian. The parameter a determines the width of the Gaussian function, the parameter b shifts it up and down. We consider $\mathbf{a \in [2,8]}$.and $\mathbf{b \in (-1,1)}$. To explore the dynamics we first plot the function and it's fixed points (intersections with the line f(x)=x) for different parameters of b.s

```matlab
f=@(x,a,b) exp(-a*x.^2)+b; 
g=@(x) x;

x=-2:0.01:2;
y1=f(x,8,1);
y2=g(x);
p1=plot(x,y1,'LineWidth',1.5); M1= 'Gaussian';
hold on;
p2=plot(x,y2,'LineWidth',1.5); M2= 'f(x)=x';
set(gca,'Ylim',[0,2])
grid on
hold on;
P=InterX([x;y1],[x;y2]);
p3=scatter(1,1,30,'ro'); M3= 'fixed point';
xlabel('x','FontSize',20);
ylabel('y','FontSize',20);
set(gca,'FontSize',16);
t=title(sprintf('One dimensional function f, a=8,b=1'));
set(t,'FontSize',16);
legend([p1;p2;p3], M1, M2,M3,'Location','southwest')
```

#### We see that for b > 0 we only get one fixed point. We expect to get three fixed points if b is sufficiently small. If be gets very small the function gets shifted down so much that we again only get one fixed point. To observe this in a plot, we consider different values of b.

```matlab
figure()
x=-1:0.01:1;
y1=f(x,8,-0.7);
y2=g(x);
p1=plot(x,y1,'LineWidth',1.5); M1= 'Gaussian';
hold on;
p2=plot(x,y2,'LineWidth',1.5); M2= 'f(x)=x';
set(gca,'Ylim',[-1,1])
grid on
P=InterX([x;y1],[x;y2]);
p3=scatter(P(1,:),P(2,:),30,'ro'); M3= 'fixed point';
xlabel('x','FontSize',20);
ylabel('y','FontSize',20);
set(gca,'FontSize',16);
t=title(sprintf('One dimensional function f, a=8,b=-0.7'));
set(t,'FontSize',16);
legend([p1;p2;p3], M1, M2,M3,'Location','northwest')
figure()
x=-1.5:0.01:1.5;
y1=f(x,8,-1.2);
y2=g(x);
p1=plot(x,y1,'LineWidth',1.5); M1= 'Gaussian';
hold on;
p2=plot(x,y2,'LineWidth',1.5); M2= 'f(x)=x';
set(gca,'Ylim',[-1.5,1.5])
set(gca,'Xlim',[-1.5,1.5])
grid on
P=InterX([x;y1],[x;y2]);
p3=scatter(P(1,:),P(2,:),30,'ro'); M3= 'fixed point';
xlabel('x','FontSize',20);
ylabel('y','FontSize',20);
set(gca,'FontSize',16);
t=title(sprintf('One dimensional function f, a=8,b=-1.2'));
set(t,'FontSize',16);
legend([p1;p2;p3], M1, M2,M3,'Location','northwest')
```

### Question 4

#### We consider the two dimensional map 

$$\mathbf {x_{n+1}=x_n^2-y_n^2+ax_n+by_n}$$

$$\mathbf {y_{n+1}=2x_ny_n+cx_n+dy_n}$$

#### We want to compute the attractor on the (x,y)-plane for different sets of parameters. First we consider (a=0.9,b=-0.6013,c=2,d=0.5). We run the system for one million steps and then consider the last 100,000 steps (last 10%). Plotting our results, we see that the system is attracted to a strange attractor. To compute its correlation dimension, we pick a random point x on the attractor and consider a circle with radius $\mathbf{\epsilon}$ around it. We then count how many points on the attractor are in this circle. We then slowly make epsilon bigger and count the points for each size. We repeat this process for many different x and average over all of them. In the end we plot the different epsilons against the average points we counted. On a log-log plot the graph should look linear with the slope of the best fit line being the correlation dimension.

```matlab
f=@(x,a,b,c,d) [x(1).^2-x(2).^2+a*x(1)+b*x(2);2*x(1).*x(2)+c*x(1)+d*x(2)]; 
a=0.9;
b=-0.6013;
c=2;
d=0.5;
N=1e6;

figure(1);
hold on 

x=zeros(2,N);
x(:,1)=[0.2,0.1];
for i=1:N-1                     
    x(:,i+1)=f(x(:,i),a,b,c,d);
end
x=x(:,N-0.1*N:N);
plot(x(1,:),x(2,:),'k.')
xlabel('x','FontSize',20);
ylabel('y','FontSize',20);
t=title(sprintf('Strange attractor'));
set(t,'FontSize',16);
```

```matlab
%compute the correlation dimension 
s=size(x,2);                           % get the total number of points on the attractor
eps=logspace(-4,0,20);                 % generate epsilon values
se=size(eps,2);                     
n=1e3;                                 % set how many points we want to check
attractor=zeros(2,n);                  % initialise the attractor points
for i=1:n
    r=randi(s);                        % pick random points from the attractor (center of the epsilon circles)
    a=x(:,r);
    attractor(:,i)=a;
end,
counter =zeros(1,se);                  % initialise the counter 
for k=1:n
    a=attractor(:,k);
    V=x-a;
    V=V.*V;                            % compute the distance
    V=sqrt(sum(V,1));
    for j=1:se
        number=sum(V<eps(j));          % compare the distancte to epsilon
        counter(1,j)=counter(1,j)+number; % count the points
    end
end
counter=counter./n;                    % average
```

```matlab
p1=loglog(eps,counter,'bx','MarkerSize',8,'LineWidth',2); M1= 'data';
hold on;
c=polyfit(log10(eps),log10(counter),1)                 % fit the best fit line
p2=loglog(eps,10.^(c(1)*log10(eps)+c(2)),'r-','Linewidth',2); M2= 'best fit line';
xlabel('\epsilon','FontSize',20);
ylabel('number of points','FontSize',20);
title('Number of points in the \epsilon sphere around a','FontSize',14);
legend([p1;p2], M1, M2,'Location','northwest');
```

#### Our results show that the correlation dimension of the strange attractor is $\mathbf{\sim 1.1438}$.

```matlab
function phaseplot(mu,cut)  % take mu as input, cut specifies if we're looking at long term behaviour or not

    b=0.111;                % specify the parameters as given in the question
    V1=0.889;
    beta=0.02; 
    gamma=2;
    th=2;
    k1=0.7;
    kgamma=0.1;
    k2=0.7;
    kf=8.1;

    % x(1) = c
    % x(2) = h
    % x(3) = mu
    
    % define the function
    f=@(t,x) [x(3)*kf*x(2)*(b+(V1*x(1))/(x(1)+k1))-(gamma*x(1))/(kgamma+x(1))+beta;(k2^2/(k2^2+x(1)^2)-x(2))/th;0];

    figure(1);
    clf
    hold on;

    tmax=100;  % specify the maximum time
    lim=3;     % set the limit on the x-axis
    x0zz=linspace(0,lim,4);   % specify the density of the trajectories


    for x0=x0zz
        for y0=x0zz
            [~, x] = ode45(f, [0, tmax], [x0;y0;mu]);   % use ode45 to solve the system
            if cut==0
                p=plot_dir(x(:,1),x(:,2));  % plot the trajectories if cut==0
            end
            if cut==1
                length=size(x,1);
                p=plot(x(length-10:length,1),x(length-10:length,2),'k','LineWidth',0.5);  % plot the long term behaviour 
                % if cut ==1
            end
        end
    end

end

```

```matlab
% This function works exactly the same as the one above, expect the trajectories are more dense and the plotting 
% slightly different. 

function colourlines(mu,cut)
    b=0.111; 
    V1=0.889;
    beta=0.02; 
    gamma=2;
    th=2;
    k1=0.7;
    kgamma=0.1;
    k2=0.7;
    kf=8.1;

    % x(1) = c
    % x(2) = h
    % x(3) = mu
    f=@(t,x) [x(3)*kf*x(2)*(b+(V1*x(1))/(x(1)+k1))-(gamma*x(1))/(kgamma+x(1))+beta;(k2^2/(k2^2+x(1)^2)-x(2))/th;0];

    figure(1);
    clf
    hold on;

    tmax=100;
    lim=3;
    x0zz=linspace(0,lim,30);

    for x0=x0zz
        for y0=x0zz
            [~, x] = ode45(f, [0, tmax], [x0;y0;mu]);
            if cut==0
                p=plot(x(:,1),x(:,2),'LineWidth',0.5);
            end
            if cut==1
                length=size(x,1);
                p=plot(x(int(0.9*length):length,1),x(int(0.9*length):length,2),'LineWidth',0.5);
            end
        end
    end
end

```

```matlab
function onedim(a)                       % chose the parameter value a as input

    f=@(x,a,b) exp(-a*x.^2)+b;           % define the one-dimensional function

    fprime=@(x,a) -2*a*x.*exp(-a*x.^2);  % define the derivative of the one-dimensional function

    nb=1e3;                              % maximum value for b
    b=linspace(-1,1,nb);                 % define the linspace
    b=transpose(b);                      % transpose b (comes in handy later)

    N=1e4;                               % define the maximum value for N
    x=zeros(nb,N);                       % define a matrix x with nb rows and N columns
    x(:,1)=0.2;                          
    
    for j=1:nb                           % apply the one-dim function f to every row of the matrix (f(x_n)=x_n+1)
        for i=1:N-1                     
           x(j,i+1)=f(x(j,i),a,b(j));
        end
    end

    xS=x(:,N-100:N);                     % cut the matrix, consider only the last 100 columns of the matrix N.
                                         % We do that because it usually takes some time to reach the stable state.
                                         % We only want to see the result.
    Beez=ones(size(xS)).*b;                 % Define a matrix B of the same dimensions as x 
    
    % Compute the Layapunov exponent

    y=fprime(x,a);                       
    y=abs(y);                            
    y=log(y);                           
    L=zeros(nb,1);                       
    for j=1:nb                 
       L(j)=sum(y(j,:))/N;                
    end

    % Plot the results (Orbit diagram and Lyapunov exponent)
    
    figure();
    clf;
    hold on;
    p=plot(b,L,'k','LineWidth',1);
    p=plot(b,zeros(size(b)),'b','LineWidth',1);
    grid on;
    xlabel('b','FontSize',20);
    ylabel('\lambda','FontSize',20);
    set(gca,'FontSize',16);
    t=title(sprintf('Lyapunov exponent for a=%g',a));
    set(t,'FontSize',20);
    set(gca,'Ylim',[-1.5,0.5]);

    figure();
    clf
    hold on;
    for j=1:nb
       p=plot(Beez(j,:),xS(j,:),'k.','MarkerSize', 0.1); 
    end
    grid on;
    xlabel('b','FontSize',20);
    ylabel('x','FontSize',20);
    set(gca,'FontSize',16);
    t=title(sprintf('Orbit diagram for a=%g',a));
    set(t,'FontSize',20);

end 

```
